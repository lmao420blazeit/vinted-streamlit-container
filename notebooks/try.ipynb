```python
import requests
import numpy as np
import io
from PIL import Image
import matplotlib.pyplot as plt

def get_image_bytes(image_url):
    try:
        # Download the image
        response = requests.get(image_url)
        response.raise_for_status()  # Raise an exception for bad status codes

        # Get the image content as bytes
        image_bytes = response.content

        return image_bytes
    except Exception as e:
        print("Error:", e)
        return None

# URL of the image
image_url = "https://images1.vinted.net/t/02_02325_8Anj3abbcpMaLs4cL3hb1Erq/150x210/1692089580.jpeg?s=421d33c7bdcff7781373c518f7a805cad45d1675"

# Get the image bytes
image_bytes = get_image_bytes(image_url)

print(image_bytes)
```
